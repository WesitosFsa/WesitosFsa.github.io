```python
import pandas as pd
import os

# Definir la ruta de la carpeta que contiene los archivos
ruta_carpeta = r'C:\Users\User\Desktop\Proyecto 3\Proyecto 2\Datasets'

# Obtener la lista de archivos en la carpeta
archivos = os.listdir(ruta_carpeta)

# Convertir archivos XLSX a CSV
for archivo in archivos:
    if archivo.endswith('.xlsx'):
        ruta_archivo_xlsx = os.path.join(ruta_carpeta, archivo)
        nombre_archivo_csv = os.path.splitext(archivo)[0] + '.csv'
        ruta_archivo_csv = os.path.join(ruta_carpeta, nombre_archivo_csv)
        df = pd.read_excel(ruta_archivo_xlsx)
        df.to_csv(ruta_archivo_csv, index=False)

# Actualizar la lista de archivos después de la conversión
archivos = os.listdir(ruta_carpeta)

# Crear un diccionario para almacenar los DataFrames de cada archivo
dataframes = {}

# Iterar sobre cada archivo y cargarlo en un DataFrame
for archivo in archivos:
    if archivo.endswith('.csv'):  # Solo cargar archivos CSV
        ruta_archivo = os.path.join(ruta_carpeta, archivo)
        nombre_archivo = os.path.splitext(archivo)[0]  # Obtener el nombre del archivo sin extensión
        df = pd.read_csv(ruta_archivo)
        dataframes[nombre_archivo] = df

# Limpiar cada DataFrame por separado
dataframes_limpios = {}
for nombre, df in dataframes.items():
    df_sin_nulos = df.dropna()
    df_limpio = df_sin_nulos.drop_duplicates()
    dataframes_limpios[nombre] = df_limpio

# Ahora tienes un diccionario donde cada DataFrame mantiene su integridad y está limpio
# Puedes acceder a cada DataFrame individualmente por su nombre de archivo
```
